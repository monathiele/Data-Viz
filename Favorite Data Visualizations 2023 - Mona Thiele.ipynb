```python
#### Spotify inspired radial bar chart

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randint(6000, 10000, size=12)

# Create a figure and axis object
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'), figsize=(10, 10))

# Set the theta values for each bar
theta = np.linspace(0, 2 * np.pi, len(data), endpoint=False)

# Set the width of each bar
width = 2 * np.pi / len(data)

# Plot the bars
bars = ax.bar(theta, data, width=width, bottom=0.0)

# Set the color of each bar
colors = ['#F9D5E5', '#F0B7A4', '#E6A8D7', '#B39CD0', 
          '#AED6F1', '#A2D9CE', '#F7DC6F', '#F5B041', 
          '#D7DBDD', '#D5F5E3', '#D2B4DE', '#C39BD3']
for i, bar in enumerate(bars):
    bar.set_color(colors[i])

# Set the title of the chart
ax.set_title('DeepL Users per Month', fontsize=20, fontname='Arial')

# Add the month labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax.set_xticks(theta)
ax.set_xticklabels(months, fontsize=15, fontname='Arial')

# Remove the grid lines
ax.grid(False)

# Show the chart
plt.show()
```
